```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the size of the figure before creating the plot
plt.figure(figsize=(14, 5))

# Load the 'healthexp' dataset
healthexp_dataset = sns.load_dataset('healthexp')

# Filter the dataset for specific years
selected_years = [2010, 2011, 2012, 2013, 2014, 2015]
filtered_data = healthexp_dataset.query("Year in @selected_years")

# Use sns.lineplot to create the line plot
sns.lineplot(data=filtered_data, x="Country", y="Life_Expectancy", hue="Year", palette="muted")

# Set the title of the plot
plt.title("Life Expectancy by Country Over Time")

# Show the plot
plt.show()
```
